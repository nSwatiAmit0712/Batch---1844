# KDE(Kernel Density Plot) Plot

```python
def kdePlot(var):
    fig = plt.figure(figsize=(15,4))
    ax = sns.kdeplot(df.loc[(df['Attrition']=='No'),var],color='b',shade=True,label='no Attrition')
    ax = sns.kdeplot(df.loc[(df['Attrition']=='Yes'),var],color='r',shade=True,label=' Attrition')
    plt.title('Employee Attrition with respect to {}'.format(var))
```

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X,0.80)
len(set(corr_features))
```

```python
Attrition = df.Attrition.value_counts()
sns.barplot(x=Attrition.index,y=Attrition.values)
```

from the countplot we can say that our dataset is highly imbalanced.
Hence, we have to balance the dataset first.
Here we are going to use Over sampling technique using SMOTE.
SMOTE stands for Synthetic Minority OverSampling Technique. The advantage of SMOTE is that we are not generating duplicates but rather creating synthetic datapoints that are slightly different from original datapoints.


```python
from imblearn.over_sampling import SMOTE
```
